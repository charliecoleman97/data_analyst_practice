```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8)

# Read Data

csv_location = "D:\charlie\portfolio_project1\python_part\movies.csv"
df = pd.read_csv(csv_location)
pd.set_option('display.max_rows', None)
```

```python
# Look at the data 
# Overall goal is to find a correlation between movie budget and revenue
df.head()
```

```python
# Looking for any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Sort by gross column
df.sort_values(by=['gross'], inplace=False, ascending=False).head()
```

```python
# Drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False).head()
```

```python
# Budget and revenue correlation - scatter plot budget vs gross

plt.scatter(df['budget'], df['gross'])
plt.title('Budger vs Gross Earnings')

plt.xlabel('Budget')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
df.head()
```

```python
# PLot budget vs grsoss using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'} )
```

```python
# Looking at correlation
df.corr(method='pearson')
# df.corr(method='kendall')
# df.corr(method='spearman') # pearson, kendall, spearman
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Looking at company
df_numarised = df

for col_name in df_numarised.columns:
    if(df_numarised[col_name].dtype == 'object'):
        df_numarised[col_name] = df_numarised[col_name].astype('category')
        df_numarised[col_name] = df_numarised[col_name].cat.codes

df_numarised.head()
```

```python
correlation_matrix = df_numarised.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numarised.corr()
```

```python
correlation_mat = df_numarised.corr()
corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
